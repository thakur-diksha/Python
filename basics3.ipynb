# Sets

- unordered
- un-indexed
- un-changeable/immutable (like tuples)
### but diff bw sets and tuples is that, with sets WE CAN ADD & REMOVE ITEMS. So we cannot overwrite previously defined items in a set.
- can take diff data types
- doesn't allow duplicate elements (unlike lists and tuples)
- sets are created using curly braces {}

```python
set1 = {'123', '567', 45, 'diksha'}
set1
```

```python
set2 = {'thakur', 23, 'thakur', True}   #no duplicate elements
set2
```

### len

```python
print(len(set1))
```

### creating set using constructor

```python
st = set(('ab', 5, False))
st
```

##accessing items
### since items are unordered, we cannot access them using index, so we can only use for loop


```python
for i in set1:
  print(i)
```

```python
for i in st:
  print(i)
```

## add method
## adding new item - not overwritng

```python
set1.add(True)
set1
```

## update method
### adding items from one set to another

```python
set1.update(st)    #i.e., set1 is updated by the value of st
set1
```

### remove method

```python
set1.remove(45)   #deletes 45
set1
```

### discard method

```python
set1.discard('567')
set1
```

## pop method
### removes an item from the set

```python
set1
```

```python
set1.pop()
set1
```

```python
p = set1.pop()
print(p)
print(set1)
```

removing last element using pop

```python
s = {'oo', 'pp'}
z = s.pop()
print(z)
print(s)
```

```python
s1 = {34, 56}
z = s1.pop()
print(z)
print(s1)
```

### clear method
removes elements, but the empty set remains

```python
set1.clear()
set1
```

### del
deletes the whole set along with the items

```python
del set1
```
