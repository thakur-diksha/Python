# for loop 

for i in range(1,20,1): --> here the rightmost 1 is increment value
print(i)                 => 1 2 3 ... 19

for i in range(20,1, -2): --> here -2 is decrement value
print(i)                   => 20 18 16 ...... 2

- all the statements within a loop have to be indented. loops allow multiple statements, as long as the same indentation is maintained

- prints from (lower range) to (upper range-1)

- when no increment/decrement value is provided, result is incremented by 1, by default

### Creating a list

```python
list1 = ["diksha", "abc", "xyz"]
list1
```

### negative indexing

```python
print(list1[-1])
```

### accessing a range

```python
print(list1[0:2])  #ele at 0 and (2-1=) 1
```

### printing in reverse order 

```python
print(list1[: :-1])
```

### checking if an ele is present in list

```python
if "abc" in list1:
  print("present")
```

### changing items in list

```python
list1[2]='thakur'
list1
```

### create a list using a list constructor

```python
list2 = list(('here', 'cat', 5, 7.8))
list2
```

### changing more than one value at the same time

```python
 list2[1:3]=['dt','xvc']  #changes items at index 1(i.e, 'cat') to 'dt' and index(3-1=2)(i.e, '5') to 'xvc'
 list2
```

### inserting item in list

```python
 list2.insert(2, 'th') #inserts at index 2 pos 3, i.e., before 'xvc'
 list2
```

### append: inserts item at the end of the list

```python
 list3 = ['cycle', 'dog', 'cart', 'sweater']
 list3
```

```python
list3.append('noodles')
list3
```

## diff bw append and insert:
- with insert, we have to specify the position at which we want to insert. 
- with append, it , by default, inserts at the end of list.

### extend

```python
list1.extend(list2) #copies the value of list2 in list1 and list1 values remain as it is.
list1
```

### remove
- give the element that you want removed, not the index
- if there's a duplicate element, the first one will be removed

```python
list3.remove("noodles")
list3
```

```python
list3.append("dog")
list3
```

```python
list3.remove("dog")
list3
```

### pop

```python
list3.pop()  #removes the last element by default
```

```python
list3
```

### del

```python
del list1[0]  #removes the item whose index is provided
list1
```

## Diff bw pop, remove and del:
- with remove, item name has to be provided
- pop simply deletes the last item from the list
- del takes the index value and deletes that element

## deleting the whole list

```python
del list2
```

## clear() method

```python
list3.clear()
list3     #shows the empty list
```

## diff bw del and clear:
- del deletes the whole list along with the items
- clear only removes the items within a list, the empty list that was created is still there
