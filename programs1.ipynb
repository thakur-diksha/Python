### 5. WAP to print the Electricity Bill
### Upto 200 - 0.5/unit
### 201 - 500 - 1/unit for units consumed above 200
### 501 - 1000 - 2.5/unit for units consumed above 500
### 1001 - 1500 - 3.5/unit for units consumed above 1000
### 1501 - 2500 - 5/unit for units consumed above 1500
### Above 2500 - 10/unit for units consumed above 2500

```python
unit=float(input("enter units consumed:"))
if (unit <= 200):
  bill=float(unit * 0.5)
elif (unit <= 500):
   bill=float((200 * 0.5) + (unit - 200) * 1)
elif (unit <= 1000):
  bill=float((200 * 0.5) +(300 * 1) +(unit - 500) * 2.5)   
elif (unit <= 1500):
  bill=float((200 * 0.5) +(300 * 1) +(500 * 2.5) +(unit - 1000) * 3.5)
elif(unit <= 2500):
  bill=float((200 * 0.5) +(300 * 1) +(500 * 2.5) + (500 * 3.5) + (unit-1500) * 5)
else:
  bill=float((200 * 0.5) +(300 * 1) +(500 * 2.5) + (500 * 3.5) + (1000 * 5) + (unit - 2500) * 10 )

print("total electricity bill is", bill)
```
