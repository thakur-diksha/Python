# dictionary

- key-value pair (key:value)
- curly braces, like sets
- value/key can be any data type
- oredered (although on previous versions, dict was unordered)
- doesn't allow duplicate values - for key. allowed for value
- changeable


### creating dictionary

```python
stu1 = {"Name":"Diksha", "Branch":"IT", "Age":20}
stu1
```

```python
stu1 = {"Name":"Diksha", "Branch":"IT", "Age":20, "Age":21}  #last age is considered, duplicates not allowed for key
stu1
  
```

```python
stu2 = {"Name":"Diksha", "Branch":"IT", "Age":20, "Age1":20}   #duplicates allowed for values
stu2
```

## accessing items 
### using key name

```python
key2 = stu1["Branch"]
key2
```

### get method

```python
key2value = stu1.get("Branch")
key2value
```

### getting all the key values

```python
keyvalues = stu1.keys()
keyvalues
```

### changing key values

```python
stu1["Name"] = "Thakur"
stu1
```

### update dictionary

```python
stu1.update({"City":"Jaipur"})  #inserts a new key-value pair
stu1
```

```python
stu1["Paper"] = "AI and ML"   #new value is inserted at last
stu1
```

### changing an existing value

```python
stu1.update({"City":"Lucknow"})
stu1
```

### pop

```python
stu1.pop("Paper")    #removing items. pass key as parameter
stu1
```

### pop item

```python
stu1.popitem()   #pops the last item, by default. no parameter to be passed
stu1
```

### del

```python
del stu1["Age"]      #deletion using key as parameter
stu1
```

### deleting the whole dictionary

```python
del stu2
```

### clear

```python
stu2 = {"Name":"Diksha", "Branch":"IT", "Age":20}
```

```python
stu2.clear()
stu2
```

## diff bw del and clear
- del deletes the entire dictionary along with the items
- whereas clear method only deletes all the items, the empty dictionary exists

```python
dict = {"Name":"Diksha", "Branch":"IT", "Age":20}
dict
```

### accessing values

```python
for i in dict:
  print(dict[i])
```

```python
for i in dict.values():
  print(i)
```

### accessing keys

```python
for i in dict.keys():
  print(i)
```

### accessing both key and values

```python
for i,j in dict.items():
  print(i,j)
```

### copy method

```python
dict2 = dict.copy()
dict2
```

### multiple dictionaries simultaneously

```python
myitclass = {"student1":{"Name":"Diksha", "Class":"IT","City":"Jaipur"}, "student2":{"Name":"Thakur","Class":"IT","City":"Lko"}}
myitclass
```
