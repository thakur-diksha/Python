# FILE HANDLING

### create a new file, or open an existing file to perform read/write/modify methods on the file
- x = create
- t = text mode (if t is not given, file opens in default mode - text and read mode)
- a = append mode. appends content in file
- w =write mode. overwrites the previous content
- if a file has not already been created, but we try to open it in w/x/a mode the file is created then

without the OS module, no file can be deleted (remove method) --> import os

### creating a file named 'myfile'

```python
file1 = open("myfile.txt", "x")
print(file1)
```

### opening file in write mode

```python
file1 = open("myfile.txt","w")
file1.write("I am diksha. this is ML lab.")
file1.close()
```

### reading the file, open it in read mode first

```python
file1 = open("myfile.txt")   #if no mode specified, by default it's in text mode
print(file1.read())
```

### append mode

```python
file1 = open("myfile.txt","a")  
print(file1.write("My branch is IT"))
file1.close()
```

## deleting a file
### import the os module first

```python
import os
os.remove("myfile4.txt")
```

### checking if the file myfile4 has been deleted

```python
import os

if(os.path.exists("myfile4.txt")):
  os.remove("myfile4.txt")        #if the file still exists, delete it
else:
  print("doesn't exist.")         #else, print the msg that the file doesn't exist
```
