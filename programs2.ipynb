### 2. WAP to check if a number is prime or not

### 7. WAP to check if a number is armstrong number or not

```python
n = int(input("enter the number: ")) 
sum = 0
temp = n
while(temp > 0):
   x = temp % 10
   sum += x ** 3
   temp //= 10
if n == sum:
   print(n,"is an armstrong no")
else:
   print(n,"is not an armstrong no")
```

### 9. WAP to check if a number is digit palindrome or not

```python
n = int(input("enter a number:"))  
temp = n  
rev = 0  
while(n > 0):  
    x = n % 10  
    rev = rev * 10 + x 
    n = n // 10  
if(temp == rev):  
    print(temp, "is a palindrome number")  
else:  
    print(temp, "is not a palindrome number!")  
```

### 10. WAP to find factorial of a given number

```python
n = int(input("enter number: "))
fact = 1
if(n < 0):
  print("enter a positive number only!")
elif(n == 0):
  print("factorial of 0 is 1")
else:
  for i in range(1, n+1):
    fact = fact * i
  print("factorial of", n, "is", fact)
```

### 12. WAP to print pascal's triangle

```python
r = int(input("Enter the no of rows: "))
for i in range(r+1):
  for j in range(r-i):
    print(' ', end='')
  C = 1
  for j in range(1, i+1):
    print(C, ' ', sep='', end='')
    C = C * (i - j) // j
  print()
```

### 15. WAP to calculate roots of the quadratic equation

```python
import math
a = float(input("enter var 1 or a: "))  
b = float(input("enter var 2 or b: "))  
c = float(input("enter constant or c: "))  
if(a == 0):
  print("not a quadratic eqn. enter a value other than 0")
else: 
  d = b**2 - 4*a*c 
  sqd = math.sqrt(abs(d))

  if(d > 0):
    print("roots are real and different and they are: ") 
    print((-b + sqd)/(2 * a))  
    print((-b - sqd)/(2 * a))
  elif(d == 0): 
    print("roots are real and same and they are; ")
    print(-b / (2 * a)) 
  else:
    print("roots are complex and they are: ")
    print(- b / (2 * a), " + i", sqd)
    print(- b / (2 * a), " - i", sqd) 
```
