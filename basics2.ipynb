# Tuples

- ordered
- indexing starts from 0, just like list
- unchnageable (unlike list) - we cannot change, insert or remove items after the tuple has been created
- for doing these changes, we've to first convert the tuple into a list and then make changes
- takes any data type. one tuple can have elements with multiple DTs


## creating a tuple 
- for list: square brackets []
- for tuple: parentheses ()

```python
tuple1 = ('abc', 'xyz')
tuple1
```

### len function

```python
print(len(tuple1))
```

### indexing

```python
print(tuple1[0])
```

### takes any DT

```python
tuple2 = ('diksha', 5, 8.6)
tuple2
```

### creating tuple using tuple constructor

```python
tuple3 = tuple(("thakur", "cat", "blanket", 70))
tuple3
```

## Indexing - can be positive and negative, just like lists
## range is also the same as list

### changing tuple items
- can't be done directly since tuples are immutable
- convert to list then make changes

```python
x = list(tuple2)    #converting tuple2 to a list named x
x
```

```python
x[1] = 'mango'    #item '5' is replcaed by 'mango'
x
```

```python
x.append('98')
x
```

```python
x.pop()
x
```

```python
x.insert(2, 'horse')   #inserts 'horse' before '8.6'
x
```

```python
x.remove('mango')
x
```
